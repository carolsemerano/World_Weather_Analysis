```python
#Import dependencies
import random
import numpy as np
import requests
import os
import time
import pandas as pd

# Import citipy module
from citipy import citipy

# Import OpenWeatherMap key
from config import weather_api_key
```

```python
# Generate ramdom latitude and longitude values
lats = np.random.uniform(low=-90.000, high=90.000, size=2000)
lngs = np.random.uniform(low=-180.000, high=180.000, size=2000)
lat_lngs = zip(lats, lngs)
lat_lngs
```

```python
# Create a list with the coordinates
coordinates = list(lat_lngs)
```

```python
# Create a list for cities.
cities = []
# Get nearest city for each latitude and longitude.
for coordinate in coordinates:
    city = citipy.nearest_city(coordinate[0], coordinate[1]).city_name

    # Add city, if the it is not in the list
    if city not in cities:
        cities.append(city)
# Print the city count to confirm sufficient count.
len(cities)
```

```python
# Call API OpenWeatherMap to retrieve following info
# Latitude, longitude, Maximum temperature, Percent humidity, Percent cloudiness, 
# Wind speed, Weather description (for example, clouds, fog, light rain, clear sky)

# OpenWeatherMap API url
url = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID=" + weather_api_key
```

```python
# Create an empty array for data.
city_data = []

# Start Logging
print("Beginning Data Retrieval \n -----------------------------")

# Create counters.
record_count = 1
set_count = 1

# Loop through the cities in the list.
for i, city in enumerate(cities):

    # Group cities in sets of 50.
    if (i % 50 == 0 and i >= 50):
        set_count += 1
        record_count = 1

        # Wait 1 minute to restrict the number of calls
        time.sleep(60)

    # Create end URL
    city_url = url + "&q=" + city.replace(" ","+")

    # Log the URL, record, and set numbers and the city.
    print(f'Processing Record {record_count} of Set {set_count} | {city}')

    # Add 1 to the record count.
    record_count += 1

    # Make the request using the city_url
    try:
        # Retrieve data as JSON
        city_weather = requests.get(city_url).json()

        # Save data into variables
        lat = city_weather["coord"]["lat"]
        lon = city_weather["coord"]["lon"]
        max_temp = city_weather["main"]["temp_max"]
        humidity = city_weather["main"]["humidity"]
        clouds = city_weather["clouds"]["all"]
        wind = city_weather["wind"]["speed"]
        country = city_weather["sys"]["country"]
        curr_description = city_weather["weather"][0]["description"]

        # Add city info into city_data dictionary.
        city_data.append({"City": city.title(),
                          "Country": country,
                          "Lat": lat,
                          "Lng": lon,
                          "Max Temp": max_temp,
                          "Humidity": humidity,
                          "Cloudiness": clouds,
                          "Wind Speed": wind,
                          "Current Description": curr_description})

    # If an error is experienced, skip the city.
    except:
        print("City not found. Skipping...")
        pass

# Logging Footer.
print("----------------------------- \n Data Retrieval Complete \n-----------------------------")
```

```python
# Convert the array into DataFrame.
city_data_df = pd.DataFrame(city_data)
city_data_df.head(10)
```

```python
# Create a CSV file.
output_file = " WeatherPy_Database.csv"

# Save the data frame into the CSV file.
city_data_df.to_csv(output_file, index_label="City_ID")
```
